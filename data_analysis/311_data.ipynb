```python
%matplotlib inline
import pandas as pd

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

orig_data = pd.read_csv('./311_Service_Requests_from_2010_to_Present.csv', nrows=5000000, parse_dates=['Created Date'])
```

### Get all the different kinds of noise complaints and filter rows

```python
complaints = orig_data['Complaint Type'].unique()
noise_types = [complaint for complaint in complaints if 'noise' in complaint.lower()]
noise_types
noise_complaints = orig_data[orig_data['Complaint Type'].isin(noise_types)]
```

```python
noise_complaints['Created Date'] = noise_complaints['Created Date'].map(lambda x: x.strftime('20%y-%m-%d'))
noise_complaints['Created Date'] = pd.to_datetime(noise_complaints['Created Date'])
```

```python
noise_complaints = noise_complaints.set_index('Created Date')
```

```python
sliced = noise_complaints[['Complaint Type']]
sliced_all = sliced.ix['20120101':'20141026']
sliced_2014 = sliced.ix['20140101':'20141026']
sliced_2013 = sliced.ix['20130101':'20131231']
sliced_2012 = sliced.ix['20120101':'20121231']
```

```python
grouped_all = sliced_all.groupby(sliced_all.index)
grouped_2014 = sliced_2014.groupby(sliced_2014.index)
grouped_2013 = sliced_2013.groupby(sliced_2013.index)
grouped_2012 = sliced_2012.groupby(sliced_2012.index)
```

```python
# grouped.count().to_csv('2014_noise_counts.csv')
```

```python
grouped_2014.count().plot(figsize=(16, 10))
grouped_2013.count().plot(figsize=(16, 10))
```

```python
grouped_2012.count().plot(figsize=(16, 10))
grouped_all.count().plot(figsize=(16, 10))
```

```python
grouped_all.count().to_csv('all_noise_counts.csv')
```
